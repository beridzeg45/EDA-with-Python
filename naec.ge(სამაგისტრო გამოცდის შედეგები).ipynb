# მოსალოდნელი საგრანტო ქულა, თუკი ანალიტიკური წერის ნაწილში  მივიღებდი სხვა ქულას

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from xgboost import XGBRegressor
```

```python
data=df.loc[:,df.columns.isin(['ანალიტიკური წერა','ნორმირებული ქულა.1'])]
X=data.loc[:,data.columns.isin(['ნორმირებული ქულა.1'])==False]
Y=data.loc[:,data.columns.isin(['ნორმირებული ქულა.1'])==True]
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
```

```python
model=XGBRegressor()
```

```python
model.fit(x_train,y_train)
```

```python
testing_data_prediction=model.predict(x_test)
```

```python
plt.figure(figsize=(15,5))
plt.scatter(y_test,testing_data_prediction)
plt.title('სატესტო მონაცემების სიზუსტის ვიზუალიზაცია',size=15,fontweight='bold')
plt.xlabel('რეალური ')
plt.ylabel('მოდელის მიერ ნავარაუდევი')
plt.show()

score_1=metrics.r2_score(y_test,testing_data_prediction)
print('R square error: ',score_1)
```
